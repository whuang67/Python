```python
import findspark
findspark.init('/home/mint/spark-2.1.0-bin-hadoop2.7')
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('logistic').getOrCreate()
data = spark.read.format('libsvm').load('sample_libsvm_data.txt')
data.show()
```

```python
from pyspark.ml.classification import LogisticRegression
reg_model = LogisticRegression().fit(data)
```

```python
summary = reg_model.summary
summary.predictions.show()
summary.predictions.printSchema()
```

```python
train_data, test_data = data.randomSplit([0.7, 0.3])
final_model = LogisticRegression().fit(train_data)
results = final_model.evaluate(test_data)
results.predictions.show()
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
my_evaluate = BinaryClassificationEvaluator()
results_roc = my_evaluate.evaluate(results.predictions)
results_roc
```
